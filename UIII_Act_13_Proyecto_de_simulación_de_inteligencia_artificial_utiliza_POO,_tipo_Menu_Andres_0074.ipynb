```python
class Linea_De_Credito:

    lista_creditos = []

    # =====================================================
    # CONSTRUCTOR
    # =====================================================

    def __init__(self, numero, limite, saldo,
                 historial, interes):

        self.numero = numero
        self.limite = limite
        self.saldo = saldo
        self.historial = historial
        self.interes = interes

    # =====================================================
    # AGREGAR CRÉDITO
    # =====================================================

    def agregar_credito(self):

        Linea_De_Credito.lista_creditos.append(self)

        print("\nLínea de crédito agregada correctamente.")

    # =====================================================
    # CONSULTAR CRÉDITOS
    # =====================================================

    @classmethod
    def consultar_credito(cls):

        if cls.lista_creditos == []:

            print("\nNo hay créditos registrados.")

        else:

            print("\n================ LISTA DE CRÉDITOS ================")

            print("--------------------------------------------------------------------------")

            print("{:<15} {:<15} {:<15} {:<15} {:<15}".format(

                "CLIENTE",
                "LIMITE",
                "SALDO",
                "HISTORIAL",
                "INTERES"

            ))

            print("--------------------------------------------------------------------------")

            for datos in cls.lista_creditos:

                print("{:<15} {:<15} {:<15} {:<15} {:<15}".format(

                    datos.numero,
                    datos.limite,
                    datos.saldo,
                    datos.historial,
                    str(datos.interes) + "%"

                ))

            print("--------------------------------------------------------------------------")

    # =====================================================
    # RETIRAR CRÉDITO
    # =====================================================

    @classmethod
    def retirar_credito(cls):

        buscar = input("\nIngrese número del cliente: ")

        for datos in cls.lista_creditos:

            if datos.numero == buscar:

                retiro = float(input("Ingrese cantidad a retirar: "))

                if retiro <= datos.saldo:

                    datos.saldo = datos.saldo - retiro

                    print("\nRetiro realizado correctamente.")

                else:

                    print("\nSaldo insuficiente.")

    # =====================================================
    # PAGAR CRÉDITO
    # =====================================================

    @classmethod
    def pagar_credito(cls):

        buscar = input("\nIngrese número del cliente: ")

        for datos in cls.lista_creditos:

            if datos.numero == buscar:

                pago = float(input("Ingrese cantidad a pagar: "))

                datos.saldo = datos.saldo + pago

                if datos.saldo > datos.limite:

                    datos.saldo = datos.limite

                print("\nPago realizado correctamente.")

    # =====================================================
    # ELIMINAR CRÉDITO
    # =====================================================

    @classmethod
    def eliminar_credito(cls):

        buscar = input("\nIngrese número del cliente a eliminar: ")

        for datos in cls.lista_creditos:

            if datos.numero == buscar:

                cls.lista_creditos.remove(datos)

                print("\nCliente eliminado correctamente.")

# =========================================================
# MENÚ PRINCIPAL
# =========================================================

while True:

    print("\n====================================================")
    print("         IA PARA LÍNEA DE CRÉDITO")
    print("====================================================")

    print("1. Agregar línea de crédito")
    print("2. Consultar créditos")
    print("3. Retirar crédito")
    print("4. Pagar crédito")
    print("5. Eliminar crédito")
    print("6. Salir")

    opcion = input("\nSeleccione una opción: ")

    # =====================================================
    # OPCIONES
    # =====================================================

    if opcion == "1":

        numero = input("\nIngrese número del cliente: ")
        limite = float(input("Ingrese límite de crédito: "))
        saldo = float(input("Ingrese saldo disponible: "))
        historial = input("Ingrese historial de pagos: ")
        interes = float(input("Ingrese interés mensual: "))

        obj = Linea_De_Credito(

            numero,
            limite,
            saldo,
            historial,
            interes

        )

        obj.agregar_credito()

    elif opcion == "2":

        Linea_De_Credito.consultar_credito()

    elif opcion == "3":

        Linea_De_Credito.retirar_credito()

    elif opcion == "4":

        Linea_De_Credito.pagar_credito()

    elif opcion == "5":

        Linea_De_Credito.eliminar_credito()

    elif opcion == "6":

        print("\nPrograma finalizado.")
        break

    else:

        print("\nOpción incorrecta.")
print("PROGRAMA REALIZADO POR JAQUEZ ANDRES 0074 NL = 31")
```
